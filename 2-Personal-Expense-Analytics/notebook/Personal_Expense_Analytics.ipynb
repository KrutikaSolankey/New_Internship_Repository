```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("finance_data.csv")
print(data.head())
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
total = data['amount'].sum()
print("Total Expense =", total)
```

```python
category_expense = data.groupby('category')['amount'].sum()
print(category_expense)
```

```python
category_expense.plot(kind='bar')
plt.title("Category Wise Expenses")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```

```python
category_expense.plot(kind='pie',
                      autopct='%1.1f%%')
plt.title("Expense Distribution")
plt.ylabel('')
plt.show()
```

```python
highest = category_expense.idxmax()
print("Highest Spending Category =", highest)
```

```python
lowest = category_expense.idxmin()
print("Lowest Spending Category =", lowest)
```
